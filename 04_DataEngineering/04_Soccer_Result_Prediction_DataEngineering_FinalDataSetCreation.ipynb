# Soccer Analytics -  Predicting Soccer Match Results to Improve Chances of Winning Bets 

### Inorder to prepare the final dataset for prediction, we brainstormed and finalized on creating our dataset in the following manner:
- Consider previous three seasons of data - last two season as training data and current season for the test.
- A weighted average of previous two season team related metrics with more weightage on the recent season. For example if Chelsea has played in 2017 season then its teamwise metrics were weighted average of their performance in 2015 and 2016 season with more weightage given to 2016 and likewise for next seasons.
- Last Year and Last to Last Year fields were derived from ingame match stats for the current season.
- Specially Engineered Variables:
 - __Finish Flag of Previous Year__ - Based on whether the team played in Champions League, FA Cup or Europa League.
 - __Key Players__ - 30 Top Players per each season were identified and grouped accordingly.

#### 2. Calculate the weighted average of all team related metrics based on previous two seasons

```python
lst = list(understats_df.columns)
lst.remove('team')
lst.remove('season')
lst.remove('team_id')
cal_yr =[2016,2017,2018]
#cal_yr=[2018]
aggregrated_df=pd.DataFrame()
for i in range(0,len(cal_yr)):
    print(cal_yr[i]-1)
    df1=understats_df[(understats_df['season']==cal_yr[i]-1)]
    df2=understats_df[(understats_df['season']==cal_yr[i]-2)]
    df3=understats_df[(understats_df['season']==cal_yr[i])]
    findf=df3[['team', 'season', 'team_id']]
    findf['idx_id']=findf['team_id']
    df1['idx_id']=df1['team_id']
    df2['idx_id']=df2['team_id']
    findf.set_index('idx_id',inplace=True)
    df1.set_index('idx_id',inplace=True)
    df2.set_index('idx_id',inplace=True)
    for j in range(0,len(lst)):
        findf[lst[j]]=((df1[lst[j]]*2)+df2[lst[j]])/2
    aggregrated_df=pd.concat([aggregrated_df,findf])
```
